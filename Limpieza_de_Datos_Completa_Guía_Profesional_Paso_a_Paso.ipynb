##**üß≠ 1Ô∏è‚É£ Revisi√≥n inicial y diagn√≥stico del dataset**

Antes de limpiar, diagnostica los posibles problemas.

üîπ Objetivo: identificar tipos de datos, valores nulos, duplicados, y columnas con errores o inconsistencias.

```python
peliculas.info()
peliculas.describe(include="all")
peliculas.isnull().sum()
peliculas.duplicated().sum()
peliculas.head()
```

##**üßπ 2Ô∏è‚É£ Normalizaci√≥n de nombres de columnas**

Estandariza los nombres para que sean consistentes y f√°ciles de usar:

```python
peliculas.columns = peliculas.columns.str.lower().str.strip().str.replace(" ", "_")
peliculas.head()
```

üëâ Resultado: ['titulo', 'genero', 'a√±o', 'duracion', 'puntaje', 'director']

##**üß© 3Ô∏è‚É£ Limpieza de valores faltantes (NaN)**
üí° Tip: siempre justifica si imputas o eliminas datos; depende del impacto en el an√°lisis.

```python
# Ver distribuci√≥n de nulos
print(peliculas.isna().sum())

# Opci√≥n 1: eliminar filas si el porcentaje de nulos es bajo
peliculas = peliculas.dropna(subset=["genero"])

# Opci√≥n 2: imputar valores faltantes
peliculas["puntaje"].fillna(peliculas["puntaje"].mean(), inplace=True)
```

##**üîÑ 4Ô∏è‚É£ Eliminaci√≥n de duplicados**
üí° √ötil cuando hay registros repetidos, por ejemplo, varias filas con el mismo t√≠tulo y a√±o.

```python
peliculas = peliculas.drop_duplicates()
```

##**üß† 5Ô∏è‚É£ Correcci√≥n de errores tipogr√°ficos o inconsistencias**

Un cl√°sico: diferencias entre "Animacion" y "Animaci√≥n" o "Ciencia ficcion" y "Ciencia ficci√≥n".

üí° Objetivo: uniformar categor√≠as para evitar errores en an√°lisis o visualizaciones.

```python
peliculas["genero"] = peliculas["genero"].str.normalize('NFKD')\
                                         .str.encode('ascii', errors='ignore')\
                                         .str.decode('utf-8')\
                                         .str.lower()\
                                         .str.strip()

# Normalizar valores manualmente
peliculas["genero"].replace({
    "animacion": "animaci√≥n",
    "ciencia ficcion": "ciencia ficci√≥n",
    "accion": "acci√≥n"
}, inplace=True)
```

##**‚è±Ô∏è 6Ô∏è‚É£ Verificaci√≥n y correcci√≥n de tipos de datos**
üí° Garantiza que las operaciones estad√≠sticas o gr√°ficas funcionen correctamente.

```python
peliculas["a√±o"] = peliculas["a√±o"].astype(int)
peliculas["duracion"] = peliculas["duracion"].astype(int)
peliculas["puntaje"] = peliculas["puntaje"].astype(float)
```

##**üßÆ 7Ô∏è‚É£ Detecci√≥n y manejo de valores at√≠picos (outliers)**
üí° Esto evita sesgos en promedios o distribuciones.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=peliculas["duracion"])
plt.show()

# Filtrar outliers extremos
q1 = peliculas["duracion"].quantile(0.25)
q3 = peliculas["duracion"].quantile(0.75)
iqr = q3 - q1
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

peliculas = peliculas[(peliculas["duracion"] >= limite_inferior) & (peliculas["duracion"] <= limite_superior)]
```

##**üìÜ 8Ô∏è‚É£ Estandarizaci√≥n de formatos de fecha o tiempo (si aplica)**

Aunque en tu dataset no hay fechas, si tuvieras columnas tipo "fecha_estreno", aplica:

```python
peliculas["fecha_estreno"] = pd.to_datetime(peliculas["fecha_estreno"], errors="coerce")
```

##**üßæ 9Ô∏è‚É£ Validaci√≥n de consistencia l√≥gica**

Ejemplo: verificar que los a√±os de las pel√≠culas est√©n en un rango realista.

üí° Tambi√©n puedes validar que los puntajes est√©n entre 0 y 10.

```python
peliculas = peliculas[(peliculas["a√±o"] >= 1900) & (peliculas["a√±o"] <= 2025)]
```

##**üîç üîü Limpieza de texto y formato uniforme**
üí° Mejora la presentaci√≥n y evita duplicidades causadas por espacios o may√∫sculas.

```python
peliculas["titulo"] = peliculas["titulo"].str.strip().str.title()
peliculas["director"] = peliculas["director"].str.strip().str.title()
```

##**üß± 1Ô∏è‚É£ 1Ô∏è‚É£ Creaci√≥n de nuevas variables limpias (feature engineering b√°sico)**

Ejemplo: crear una categor√≠a de duraci√≥n.
üí° Muy √∫til para segmentar o visualizar mejor los datos.

```python
def clasificar_duracion(x):
    if x < 90: return "Corta"
    elif x <= 120: return "Media"
    else: return "Larga"

peliculas["tipo_duracion"] = peliculas["duracion"].apply(clasificar_duracion)
```

##**‚úÖ 1Ô∏è‚É£2Ô∏è‚É£ Validaci√≥n final**

üí° Aseg√∫rate de que no queden valores nulos, tipos incorrectos o duplicados.

```python
print(peliculas.info())
print(peliculas.head())
peliculas.describe()
```
